```python
import pickle

with open('metrics.pickle', 'rb') as f:
    x = pickle.load(f)
```

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
mean_map = {}
mean_recall = {}

for c in x['map']:
    mean_map[c] = np.mean(x['map'][c])
    mean_recall[c] = np.mean(x['recall'][c])
```

```python
plt.plot(mean_recall.keys(), mean_recall.values(), label='recall')
plt.plot(mean_map.keys(), mean_map.values(), label='map')
plt.xlabel('Threshold')
plt.legend()
plt.grid()
plt.vlines(0.125, 0, 0.2)
```

```python
mean_n_preds = {}

for c in x['mean_n_predictions']:
    mean_n_preds[c] = np.mean(x['mean_n_predictions'][c])

plt.plot(mean_n_preds.keys(), mean_n_preds.values())
plt.grid()
plt.ylabel('Mean number of predictions')
plt.xlabel('Threshold')
plt.vlines(0.125, 0, 50)
```
